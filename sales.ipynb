                                               Analysis using plots                                                                   

1.Sales by Product

```python
df.groupby("Product")["Sales"].sum().plot(kind="bar")
plt.title("Total Sales by Product")
plt.ylabel("Sales")
plt.show()
```

Observation:

--> Headphones generate the highest sales, followed by Tablet and Laptop.

--> Camera shows moderate performance, while Mobile has the lowest total sales.

--> Overall, sales are dominated by audio and portable device categories.

2.Price vs Sales

```python
plt.scatter(df["Price"], df["Sales"])
plt.title("Price vs Sales Scatter Plot")
plt.xlabel("Price")
plt.ylabel("Sales")
plt.show()
```

Observations

--> There is a clear positive correlation between Price and Sales  as the price increases, the sales also tend to increase.

--> Data points spread widely but follow an upward trend, indicating that higher-priced products generate higher sales revenue.

--> The plot shows some variability, meaning sales are not uniform for the same price levels  different products at the same price point have different sales outcomes.

--> At lower price ranges (010,000), sales are mostly clustered near lower sales values.

-->As prices move toward higher ranges (30,00050,000), the spread in sales values increases, showing greater variation in high-priced product sales performance.

--> A few points indicate very high sales, suggesting some premium products perform exceptionally well.

3.Sales by Region

```python
df.groupby("Region")["Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales Share by Region")
plt.ylabel("")
plt.show()
```

Observations

--> West region contributes the highest share, accounting for 29.1% of total sales.

--> North and South regions have equal contributions, each making up 25.9%, indicating balanced performance between them.

--> East region has the lowest sales share, contributing only 19.1%.

--> The distribution suggests that West is the leading market, while East may need improvement to match the other regions.

--> The chart shows a moderately balanced distribution overall, except for East which lags behind.

                                                   Final Analysis                                                                          

--> West region has the highest sales share, while East contributes the least.

--> Sales increase with price, showing a clear positive correlation despite some variability.

--> Headphones lead in total sales, followed by Tablet and Laptop, while Mobile has the lowest sales.
